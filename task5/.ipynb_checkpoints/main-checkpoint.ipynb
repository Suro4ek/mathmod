```python
import gmsh
```

```python
def create_model(width, height, radius, p_reactangle, p_circle, name):
    gmsh.initialize()

    p1 = gmsh.model.geo.addPoint(0, 0, 0, p_reactangle)
    p2 = gmsh.model.geo.addPoint(width, 0, 0, p_reactangle)
    p3 = gmsh.model.geo.addPoint(width, height - radius, 0, p_circle)
    p4 = gmsh.model.geo.addPoint(width - radius, height, 0, p_circle)
    p5 = gmsh.model.geo.addPoint(width, height, 0, p_circle)
    p6 = gmsh.model.geo.addPoint(0, height, 0, p_reactangle)
        # Создаем линии прямоугольника
    l1 = gmsh.model.geo.addLine(p1, p2)
    l2 = gmsh.model.geo.addLine(p2, p3)
    l_circle = gmsh.model.geo.addCircleArc(p4, p5, p3)
    l3 = gmsh.model.geo.addLine(p4, p6)
    l4 = gmsh.model.geo.addLine(p6, p1)

        # Создаем замкнутый контур прямоугольника
    cl = gmsh.model.geo.addCurveLoop([l1, l2, l3, -l_circle, l4])
    surface = gmsh.model.geo.addPlaneSurface([cl])
    gmsh.model.addPhysicalGroup(2, [surface], 1)
    gmsh.model.geo.synchronize()
    gmsh.model.mesh.generate(2)
    gmsh.option.setNumber("Mesh.MshFileVersion", 2)
    gmsh.write(name)
    gmsh.finalize()

create_model(1, 1, 0.2, 0.05, 0.01, './models/correct.msh')
```

```python
from dolfin import *
import numpy as np
```

```python
class Solver():
    def __init__(self, name: str):
        self.__loadMesh(name)
        self.bcs = []
        self.T = TrialFunction(self.Q)
        self.v = TestFunction(self.Q)      
        
        days = 86400
        self.TMax = 30 * days
        self.dt = 0.5 * days
        self.rhoL = Constant(6.0 * 1.0e7)
        self.croS = Constant(2.0 * 1.0e6)
        self.croL = Constant(2.5 * 1.0e6)
        self.kS = Constant(1.8)
        self.kL = Constant(2.0)
        self.delta = Constant(1.0)
        
        
        initialT = Constant(2.0)
        self.T0 = interpolate(initialT, self.Q)

        self.__loadBoundaries()

        self.__loadVelocity()
    def __loadMesh(self, name: str):
        self.mesh = Mesh(f'{name}.xml')
        self.subdomains = MeshFunction('size_t', self.mesh, f'{name}_physical_region.xml')
        self.boundaries = MeshFunction('size_t', self.mesh, f'{name}_facet_region.xml')
        self.Q = FunctionSpace(self.mesh, "CG", 1)
        
        self.dx = Measure('dx', domain = self.mesh, subdomain_data = self.subdomains)
        self.ds = Measure('ds', domain = self.mesh, subdomain_data = self.boundaries)
    def __loadBoundaries(self):
        T1 = Constant(-100.0)
        bc_1 = DirichletBC(self.Q, T1, self.boundaries, 1)
        self.bcs = [bc_1]
    def __loadVelocity(self):
        phi = conditional(gt(self.T0, self.delta), 1.0, conditional(lt(self.T0, -self.delta), 0.0, (self.T0 + self.delta) / (2 * self.delta)))
        dphi = conditional(gt(self.T0, self.delta), 0.0, conditional(lt(self.T0, -self.delta), 0.0, 1.0/(2 * self.delta)))
        k = self.kS + phi * (self.kL - self.kS)
        cro = self.croS + phi * (self.croL - self.croS)
        
        self.a = (cro + self.rhoL * dphi) * (self.T / self.dt) * self.v * self.dx + inner(k * grad(self.T), grad(self.v)) * self.dx
        self.L = (cro + self.rhoL * dphi) * (self.T0 / self.dt) * self.v * self.dx 
    def solve(self, linear_solver: str, preconditioner: str):
        solutions = []
        t = self.dt
        w = Function(self.Q) 
        iterations = []
        while t < self.TMax + DOLFIN_EPS:
            t += self.dt
            A = assemble(self.a)
            b = assemble(self.L)
            [bc.apply(A, b) for bc in self.bcs]
            solver = KrylovSolver(linear_solver, preconditioner)
            parameters = solver.parameters
            num_iterations = solver.solve(A, w.vector(), b)
            iterations.append(num_iterations)
            self.T0.assign(w)
            solutions.append(w.copy(deepcopy=True))
        print(f"linear_solver={linear_solver}", f"preconditioner={preconditioner}", f"number of iterations {np.max(iterations)}")
        return solutions
```

```python
solver = Solver('./models/correct')
```

```python
solutions = solver.solve('bicgstab', 'icc')
solver.solve('bicgstab', 'petsc_amg')
solver.solve('bicgstab', 'sor')
solver.solve('cg', 'petsc_amg')
solver.solve('gmres', 'icc')
solver.solve('gmres', 'petsc_amg')
solver.solve('gmres', 'sor')
```
