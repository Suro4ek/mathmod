```python
import numpy as np
import gmsh
gmsh.initialize()
```

## Вариант 1

```python
p = 1
box_height = 20
box_length = 100
box_weight = 60
low_radius = 10//2
middle_radius = 20//2
high_radius = 30//2
gmsh.clear()
```

```python
def generate_circle(rx, ry, rz, r):
    r1 = gmsh.model.geo.add_point(rx, ry, rz, p)
    r2 = gmsh.model.geo.add_point(rx + r, ry, rz, p)
    r3 = gmsh.model.geo.add_point(rx, ry + r, rz, p)
    r4 = gmsh.model.geo.add_point(rx - r, ry, rz, p)
    r5 = gmsh.model.geo.add_point(rx, ry - r, rz, p)
    line1 = gmsh.model.geo.addCircleArc(r2, r1, r3)
    line2 = gmsh.model.geo.addCircleArc(r3, r1, r4)
    line3 = gmsh.model.geo.addCircleArc(r4, r1, r5)
    line4 = gmsh.model.geo.addCircleArc(r5, r1, r2)
    curve = gmsh.model.geo.addCurveLoop([line1, line2, line3, line4])
    return curve
```

```python
def create_box(heigth, length, weigth, mini_length, mini_weigth, low_radius):
    points_box_low = []
    points_box_low.append(gmsh.model.geo.add_point(0, 0, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(0, mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(mini_length, mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(mini_length, 2*mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(0, 2*mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(0, weigth, 0, p))

    points_box_low.append(gmsh.model.geo.add_point(length, weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(length, 2*mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(length-mini_length, 2*mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(length-mini_length, mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(length, mini_weigth, 0, p))
    points_box_low.append(gmsh.model.geo.add_point(length, 0, 0, p))
    lines = []
    for i in range(len(points_box_low)-1):
        lines.append(gmsh.model.geo.addLine(points_box_low[i], points_box_low[i+1]))
    lines.append(gmsh.model.geo.addLine(points_box_low[-1], points_box_low[0]))
    rx, ry = length//2, weigth//2
    cl_circle = generate_circle(rx, ry, 0, low_radius)
    cl = gmsh.model.geo.addCurveLoop(lines)
    s = gmsh.model.geo.add_plane_surface([cl, cl_circle])
    gmsh.model.geo.extrude([(2, s)], 0, 0, heigth, recombine=False)
```

```python
def create_cylinder(length, weigth, box_heigth, low_radius):
    rx, ry = length//2, weigth//2
    main_circle = generate_circle(rx, ry, box_heigth, high_radius)
    low_circle = generate_circle(rx, ry, box_heigth, low_radius)
    s = gmsh.model.geo.add_plane_surface([low_circle, main_circle])
    gmsh.model.geo.extrude([(2, s)], 0, 0, 30, recombine=False)
```

```python
def create_cylinder1(length, weigth, box_heigth, middle_radius):
    rx, ry = length//2, weigth//2
    main_circle = generate_circle(rx, ry, box_heigth, high_radius)
    middle_circle = generate_circle(rx, ry, box_heigth, middle_radius)
    s = gmsh.model.geo.add_plane_surface([main_circle, middle_circle])
    tag = gmsh.model.geo.extrude([(2, s)], 0, 0, 30, recombine=False)
    return tag
```

```python
def create_triagle(box_height, box_weight, weight):
    triagle_points = [
        gmsh.model.geo.add_point(15, box_weight//2-weight//2, box_height, p),
        gmsh.model.geo.add_point(35, box_weight//2-weight//2, box_height, p),
        gmsh.model.geo.add_point(35, box_weight//2-weight//2, box_height+40, p),
    ]
    line1 = gmsh.model.geo.addLine(triagle_points[0], triagle_points[1])
    line2 = gmsh.model.geo.addLine(triagle_points[1], triagle_points[2])
    line3 = gmsh.model.geo.addLine(triagle_points[2], triagle_points[0])
    curve = gmsh.model.geo.addCurveLoop([line1, line2, line3])
    s = gmsh.model.geo.add_plane_surface([curve])
    tag = gmsh.model.geo.extrude([(2, s)], 0, weight, 0, recombine=False)
    return tag
```

```python
def create_triagle1(box_height, box_weight, weight):
    triagle_points = [
        gmsh.model.geo.add_point(85, box_weight//2-weight//2, box_height, p),
        gmsh.model.geo.add_point(65, box_weight//2-weight//2, box_height, p),
        gmsh.model.geo.add_point(65, box_weight//2-weight//2, box_height+40, p),
    ]
    line1 = gmsh.model.geo.addLine(triagle_points[0], triagle_points[1])
    line2 = gmsh.model.geo.addLine(triagle_points[1], triagle_points[2])
    line3 = gmsh.model.geo.addLine(triagle_points[2], triagle_points[0])
    curve = gmsh.model.geo.addCurveLoop([line1, line2, line3])
    s = gmsh.model.geo.add_plane_surface([curve])
    gmsh.model.geo.extrude([(2, s)], 0, weight, 0, recombine=False)
    gmsh.model.add_physical_group
```

```python
create_box(box_height, box_length, box_weight, 10, 20, low_radius)
gmsh.model.geo.synchronize()
create_cylinder(box_length, box_weight, box_height, low_radius)
gmsh.model.geo.synchronize()
create_cylinder1(box_length, box_weight, box_height+30, middle_radius)
gmsh.model.geo.synchronize()
create_triagle(box_height, box_weight, 10)
gmsh.model.geo.synchronize()
create_triagle1(box_height, box_weight, 10)
gmsh.model.geo.synchronize()
gmsh.model.mesh.generate(3)
gmsh.option.setNumber("Mesh.MshFileVersion", 2)
gmsh.write("model.msh")
gmsh.finalize()
```
