Вариант 1: Прямоугольная область с круглым отверстием в центре  
Поставновка задачи 
Уравнение упругости:
$$\nabla \cdot\sigma+F=0$$
Граничные условия на $Г_D:$
$$u=e^{-\frac{x^2}{2}}cos(\pi y) {,} \ (x,y)\inГ_D$$
Граничные условия на $Г_R:$
$$\sigma\cdot n = T {,} \ T = (Tcos(\theta){,}Tsin(\theta)){,} \ \ \theta=[0,2\pi)$$
$Коэффициенты:$
$$\lambda=90 {,} \ \mu=45 {,} \ F=(2xy{,}x^2+y^2){,} \ T=5$$
где  
$Г_D$: вся внешная прямоугольная область,  
$Г_R$: весь внешний контур отверстия.

```python
import gmsh
```

```python
def generate_circle(rx, ry, rz, r, p):
    r1 = gmsh.model.geo.add_point(rx, ry, rz, p)
    r2 = gmsh.model.geo.add_point(rx + r, ry, rz, p)
    r3 = gmsh.model.geo.add_point(rx, ry + r, rz, p)
    r4 = gmsh.model.geo.add_point(rx - r, ry, rz, p)
    r5 = gmsh.model.geo.add_point(rx, ry - r, rz, p)
    line1 = gmsh.model.geo.addCircleArc(r2, r1, r3)
    line2 = gmsh.model.geo.addCircleArc(r3, r1, r4)
    line3 = gmsh.model.geo.addCircleArc(r4, r1, r5)
    line4 = gmsh.model.geo.addCircleArc(r5, r1, r2)
    curve = gmsh.model.geo.addCurveLoop([line1, line2, line3, line4])
    return curve
```

```python

def create_model(width, height, hole_radius, p_reactangle, p_circle, name):
    gmsh.initialize()

    p1 = gmsh.model.geo.addPoint(0, 0, 0, p_reactangle)
    p2 = gmsh.model.geo.addPoint(width, 0, 0, p_reactangle)
    p3 = gmsh.model.geo.addPoint(width, height, 0, p_reactangle)
    p4 = gmsh.model.geo.addPoint(0, height, 0, p_reactangle)

    # Создаем линии прямоугольника
    l1 = gmsh.model.geo.addLine(p1, p2)
    l2 = gmsh.model.geo.addLine(p2, p3)
    l3 = gmsh.model.geo.addLine(p3, p4)
    l4 = gmsh.model.geo.addLine(p4, p1)

    # Создаем замкнутый контур прямоугольника
    cl = gmsh.model.geo.addCurveLoop([l1, l2, l3, l4])
    # Создаем замкнутый контур отверстия
    cl2 = generate_circle(width / 2, height / 2, 0, hole_radius, p_circle)
    surface = gmsh.model.geo.addPlaneSurface([cl, cl2])
    gmsh.model.addPhysicalGroup(2, [surface], 1)
    gmsh.model.addPhysicalGroup(1, [cl], 1)
    gmsh.model.addPhysicalGroup(1, [cl2], 2)
    gmsh.model.geo.synchronize()
    gmsh.model.mesh.generate(2)
    gmsh.option.setNumber("Mesh.MshFileVersion", 2)
    gmsh.write(name)
    gmsh.finalize()

create_model(2, 1, 0.1, 0.1, 0.01, './models/m1.msh')
create_model(2, 1, 0.1, 0.05, 0.005, './models/m2.msh')
create_model(2, 1, 0.1, 0.02, 0.002, './models/m3.msh')
create_model(2, 1, 0.1, 0.01, 0.001, './models/correct.msh')
```

```python
from dolfin import *
```

```python
class Source(UserExpression):
    def eval(self, values, x):
        values[0] = 2 * x[0] * x[1]
        values[1] = x[0] * x[0] + x[1] * x[1]
    def value_shape(self):
        return (2,)
```

```python
class Solver():
    def __init__(self, name: str):
        self.__loadMesh(name)
        self.bcs = []

        self.u = TrialFunction(self.V)
        self.v = TestFunction(self.V)
        self.T = Constant(5)
        self.k = Constant(0.1)
        self.lamda = 90
        self.mu = 45
        self.f = Source(degree=2)
        theta1 = Expression('2 * pi * x[0]', degree = 2)
        theta2 = Expression('2 * pi * x[1]', degree = 2)
        self.ubound = Expression(('exp(-x[0]*x[0]/2)', 'cos(pi * x[1])'), degree = 2)
        self.Tbound = Expression(('Tbou * cos(theta1)', 'Tbou*sin(theta2)'), degree = 2, Tbou = self.T, theta1 = theta1, theta2 = theta2)

        self.__loadBoundaries()
    def __loadMesh(self, name: str):
        self.mesh = Mesh(f'{name}.xml')
        self.subdomains = MeshFunction('size_t', self.mesh, f'{name}_physical_region.xml')
        self.boundaries = MeshFunction('size_t', self.mesh, f'{name}_facet_region.xml')
        self.V = VectorFunctionSpace(self.mesh, 'CG', 1)
        self.dx = Measure('dx', domain = self.mesh, subdomain_data = self.subdomains)
        self.ds = Measure('ds', domain = self.mesh, subdomain_data = self.boundaries)
    def __loadBoundaries(self):
        bc_1 = DirichletBC(self.V, self.ubound, self.boundaries, 1)
        bc_2 = DirichletBC(self.V, self.Tbound, self.boundaries, 2)
        bc_3 = DirichletBC(self.V, self.Tbound, self.boundaries, 3)
        bc_4 = DirichletBC(self.V, self.Tbound, self.boundaries, 4)
        self.bcs.append(bc_1)
        self.bcs.append(bc_2)
        self.bcs.append(bc_3)
        self.bcs.append(bc_4)
    def solve(self) -> Function:
        d = self.u.geometric_dimension()
        def epsilon(u):
            return 0.5 * (nabla_grad(u) + nabla_grad(u).T)
        def sigma(u):
            return self.lamda * nabla_grad(u) * Identity(d) + 2 * self.mu * epsilon(u)
        a = inner(sigma(self.u), epsilon(self.v))*dx
        L = dot(self.f, self.v) * dx + dot(self.Tbound, self.v) * ds(5)
        u = Function(self.V)
        solve(a == L, u, self.bcs)
        return u
```

```python
solver = Solver('./models/correct')
```

```python
u_correct = solver.solve()
```

```python
solutions = []
```

```python
solver1 = Solver('./models/m1')
solver2 = Solver('./models/m2')
solver3 = Solver('./models/m3')
```

```python
u1 = solver1.solve()
u2 = solver2.solve()
u3 = solver3.solve()
```

```python
solutions.append([solver1, u1])
solutions.append([solver2, u2])
solutions.append([solver3, u3])
```

```python
errors = []
```

```python
for sol in solutions:
    solver  = sol[0]
    u = sol[1]
    u_correct_another_V = interpolate(u_correct, solver.V)
    t = inner(u - u_correct_another_V, u - u_correct_another_V)*solver.dx
    b = inner(u,u)*solver.dx
    L2 = sqrt(abs(assemble(t))/abs(assemble(b)))*100

    t = inner(grad(u - u_correct_another_V), grad(u - u_correct_another_V))*solver.dx
    b = inner(grad(u),grad(u))*solver.dx
    H1 = sqrt(abs(assemble(t))/abs(assemble(b)))*100
    errors.append([L2, H1])
```
