```python
def create_model(width, height, p_reactangle, name):
    gmsh.initialize()

    p1 = gmsh.model.geo.addPoint(0, 0, 0, p_reactangle)
    p2 = gmsh.model.geo.addPoint(width, 0, 0, p_reactangle)
    p3 = gmsh.model.geo.addPoint(width, height, 0, p_reactangle)
    p4 = gmsh.model.geo.addPoint(0, height, 0, p_reactangle)

    # Создаем линии прямоугольника
    l1 = gmsh.model.geo.addLine(p1, p2)
    l2 = gmsh.model.geo.addLine(p2, p3)
    l3 = gmsh.model.geo.addLine(p3, p4)
    l4 = gmsh.model.geo.addLine(p4, p1)

    # Создаем замкнутый контур прямоугольника
    cl = gmsh.model.geo.addCurveLoop([l1, l2, l3, l4])
    surface = gmsh.model.geo.addPlaneSurface([cl])
    gmsh.model.addPhysicalGroup(2, [surface], 1)
    gmsh.model.addPhysicalGroup(1, [l1], 2)
    gmsh.model.addPhysicalGroup(1, [l2], 3)
    gmsh.model.addPhysicalGroup(1, [l3], 4)
    gmsh.model.addPhysicalGroup(1, [l4], 1)
    gmsh.model.geo.synchronize()
    gmsh.model.mesh.generate(2)
    gmsh.option.setNumber("Mesh.MshFileVersion", 2)
    gmsh.write(name)
    gmsh.finalize()

create_model(5, 1, 0.5, './models/m1.msh')
create_model(5, 1, 0.1, './models/m2.msh')
create_model(5, 1, 0.01, './models/correct.msh')
```

```python
class Solver():
    def __init__(self, name: str):
        self.__loadMesh(name)
        self.bcs = []
        self.u, self.p = TrialFunctions(self.W)
        self.v, self.q = TestFunctions(self.W)
        
        self.dt = 100.0
        self.T = 1000.0
        self.nu = 0.001
        self.rho = 1000
        
        self.gamma = Constant(0.01)
        self.f = Constant((0, 0))
        
        self.__loadBoundaries()
        self.__loadVelocity()

    def __loadMesh(self, name: str):
        self.mesh = Mesh(f'{name}.xml')
        self.subdomains = MeshFunction('size_t', self.mesh, f'{name}_physical_region.xml')
        self.boundaries = MeshFunction('size_t', self.mesh, f'{name}_facet_region.xml')
        self.V = VectorElement("Lagrange", self.mesh.ufl_cell(), 2)
        self.Q = FiniteElement("Lagrange", self.mesh.ufl_cell(), 1)
        self.TH = self.V * self.Q
        self.W = FunctionSpace(self.mesh, self.TH)
        
        self.dx = Measure('dx', domain = self.mesh, subdomain_data = self.subdomains)
        self.ds = Measure('ds', domain = self.mesh, subdomain_data = self.boundaries)
    def __loadBoundaries(self):
        u0 = Constant((0.1, 0, 0))
        bc_1 = DirichletBC(self.W, u0, self.boundaries, 1)
        self.bcs.append(bc_1)
    def __loadVelocity(self):
        self.w = Function(self.W)
        self.u1 = Function(self.W)
        u0, p0 = split(self.w)
        self.a = self.rho / self.dt * inner(self.u, self.v) * self.dx + self.rho * inner(grad(self.u) * u0, self.v) * self.dx + self.nu * inner(grad(self.u), grad(self.v)) * self.dx + self.p * div(self.v) * self.dx + self.q * div(self.u) * self.dx
        self.L = self.rho / self.dt * inner(u0, self.v) * self.dx + inner(self.f, self.v) * self.dx
        self.A = assemble(self.a)
    def solve(self):
        solutions = []
        t = self.dt
        while t < self.T + DOLFIN_EPS:
            b1 = assemble(self.L)
            [bc.apply(self.A, b1) for bc in self.bcs]
            solve(self.A, self.u1.vector(), b1, "default", "default")
            solutions.append(self.u1.copy(deepcopy=True))
            self.w.assign(self.u1)
            t += self.dt
        return solutions
```

```python
solver = Solver('./models/correct')
```

```python
solutions = solver.solve()
```

```python
uf = File("results/velocity.pvd")
pf = File("results/pressure.pvd")
for solve in solutions:
    u1, p1 = solve.split()
    uf << u1
    pf << p1
```

```python
solver1 = Solver('./models/m1')
solver2 = Solver('./models/m2')
```

```python
from dolfin import *
```

```python
solutions1 = solver1.solve()
solutions2 = solver2.solve()
```

```python
errors = []
errors1 = []
```

```python
for i in range(len(solutions)):
    w_correct = solutions[i]
    u_correct, p_correct = w_correct.split()
    w = solutions1[i]
    u, p = w.split()
    V_space = FunctionSpace(solver.mesh, solver.V)
    u_correct_another_V = interpolate(u_correct, V_space)
    t = inner(u - u_correct_another_V, u - u_correct_another_V)*solver1.dx
    b = inner(u,u)*solver1.dx
    L2 = sqrt(abs(assemble(t))/abs(assemble(b)))*100

    t = inner(grad(u - u_correct_another_V), grad(u - u_correct_another_V))*solver1.dx
    b = inner(grad(u),grad(u))*solver1.dx
    H1 = sqrt(abs(assemble(t))/abs(assemble(b)))*100
    Q_space = FunctionSpace(solver.mesh, solver.Q)
    p_correct_another_V = interpolate(p_correct, Q_space)
    t = inner(p - p_correct_another_V, p - p_correct_another_V)*solver1.dx
    b = inner(p,p)*solver1.dx
    L2_p = sqrt(abs(assemble(t))/abs(assemble(b)))*100

    t = inner(grad(p - p_correct_another_V), grad(p - p_correct_another_V))*solver1.dx
    b = inner(grad(p),grad(p))*solver1.dx
    H1_p = sqrt(abs(assemble(t))/abs(assemble(b)))*100
    errors.append([L2, H1, L2_p, H1_p])
```
